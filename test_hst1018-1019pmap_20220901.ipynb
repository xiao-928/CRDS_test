```python
import os, re, crds
from crds import certify
from crds.certify import CertifyScript

crds_cache = "/nfsdata/users/bwzhao/data_test/cache/"
crds_cache_path = "/nfsdata/users/bwzhao/data_test/cache/mappings/hst/"
data_path = '/nfsdata/users/bwzhao/data_test/'  # the path of the *.fits files to be submitted
old_pmap = 'hst_1018.pmap'

os.environ["CRDS_CACHE"] = crds_cache
os.environ["crds_cache_path"] = crds_cache
os.environ["data_path"] = data_path
os.environ['old_pmap'] = str(old_pmap)

def generate_new_version_map(old_map):
    ### input: the name of the old version mapping file
    ### output: the name of the new version mapping file
    sp = re.split('_|\.', old_map)
    if len(sp) == 5:
        new_version = int(sp[3]) + 1
        new_map = sp[0]+'_'+sp[1]+'_'+sp[2]+'_'+str("%04d"%new_version)+'.'+sp[4]
    elif len(sp) == 3:
        new_version = int(sp[1]) + 1
        new_map = sp[0]+'_'+str("%04d"%new_version)+'.'+sp[2]
    return new_map

```

```python
old_rmap_bias = 'hst_stis_biasfile_0330.rmap'
new_rmap_bias = generate_new_version_map(old_rmap_bias)

os.environ['old_rmap_bias'] = str(old_rmap_bias)
os.environ['new_rmap_bias'] = str(new_rmap_bias)

# !crds refactor insert ./cache/mappings/hst/hst_stis_biasfile_0330.rmap hst_stis_biasfile_0331.rmap ./*_bia.fits

!crds refactor2 insert_reference \
--old-rmap $old_rmap_bias \
--new-rmap $new_rmap_bias \
--references $data_path*_bia.fits --hst --stats 

!crds refactor2 set_header --header-key name --header-value $new_rmap_bias --rmaps $data_path$new_rmap_bias --hst 
!crds checksum $data_path$new_rmap_bias
!crds diff $old_rmap_bias $data_path$new_rmap_bias --brief --mapping-text-diffs --include-header-diffs --hst
!crds certify --comparison-context=$old_pmap_bias $data_path$new_rmap_bias
```
